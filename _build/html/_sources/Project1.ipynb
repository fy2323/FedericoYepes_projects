```python
import csv
with open("Agricultural.csv", 'r', encoding='ISO-8859-1') as file:
    df = list(csv.DictReader(file))
```

```python
# List of years
years = ['1985', '1995', '2005', '2010', '2015','2018', '2019', '2020']

data_values = {}
mean_values = {}

# Populate data values for each year
for year in years:
    values = [value for value in countries[countries['Year'] == year]['Value']]
    data_values[year] = values  

# Calculate mean values for each year
for year in years:
    if data_values[year]:
        mean_values[f"mean_value_{year}"] = sum(data_values[year]) / len(data_values[year])
    else:
        mean_values[f"mean_value_{year}"] = 0 
```

```python
for year in years:
    print(f"Mean value for {year}: {mean_values[f'mean_value_{year}']:.2f}")
```

```python
# Calculate median values for each year 
median_values = {}

for year in years:
    values = data_values[year]  
    sorted_data = sorted(values)
    n = len(sorted_data)
    if n > 0:
        median_values[year] = sorted_data[n // 2] if n % 2 == 1 else (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    else:
        median_values[year] = None

# Print median values for each year
for year in years:
    print(f"Median value for {year}: {median_values[year]}")
```

```python
from statistics import multimode
mode_values = {}
# Calculate mode values for each year
for year in years:
    mode_values[year] = multimode(data_values[year])
# Print mode values for each year
for year in years:
    print(f"Mode value(s) for {year}: {mode_values[year]}")
```

```python
#Step 6 (table)
#Mean per year 
mean_per_year = countries.groupby('Year')['Value'].mean()

#Creating the df
mean_per_year_df = mean_per_year.reset_index()
mean_per_year_df.columns = ["Year", "Values"]
mean_per_year_df
```

```python
#Step 6 (sparklines)

mean_per_year_df['Sparkline'] = mean_per_year_df['Values'].apply(lambda x: '*' * round(x / 10))
mean_per_year_df
```
